```python
# Spring Climate table

# "bins" is a list of the upper limits of the bins to categorize the "Hs" (significant wave height) values.
# "dir_tot" is a list of the upper limits of the bins to categorize the wave direction values in degrees.
# "spring_table" is an empty list that will hold the values calculated for the spring season.

bins=[0.0,0.5,1.0,2.0,3.0,4.0,9.0]
dir_tot = [  0,  15,  30,  45,  60,  75,  90, 105, 120, 135, 150, 165, 180,
       195, 210, 225, 240, 255, 270, 285, 300, 315, 330, 345, 360]
spring_table = []

# This loop iterates through the dir_tot list, with i being the index for each upper limit of the bins for wave direction values.
# spe_dir is set to the data_spring array (not shown here), which contains Hs and wave direction values for the spring season.
# mm is an array of indices where the wave direction values fall within the current bin, for each row of data_spring.
# x is an array of rows from data_spring where the wave direction values fall within the current bin.
# Another loop iterates through the bins list, with j being the index for each upper limit of the bins for Hs values.
# nn is an array of indices where the Hs values fall within the current bin, for each row of x.
# y is an array of rows from x where the Hs values fall within the current bin.
# column is a list that holds the fractions of rows in y relative to the total number of rows in data_spring.
# Each column list is appended to spring_table after all bins for the current wave direction bin have been processed.

for i in range(len(dir_tot)-1):
    spe_dir = np.array(data_spring)
    mm = np.where((np.squeeze(spe_dir[:,1:]) >= dir_tot[i]) & (np.squeeze(spe_dir[:,1:]) <dir_tot[i+1]))
    x = spe_dir[np.squeeze(mm)]
    column = []
    for j in range(len(bins)-1):
        nn = np.where((np.squeeze(x[:,:1]) >= bins[j]) & (np.squeeze(x[:,:1]) <bins[j+1]))
        y = x[np.squeeze(nn)]
        column.append(len(y)/len(data_spring))
    spring_table.append(column)
    
spring_table = pd.DataFrame(np.array(spring_table))

spring_table.columns =['0.0<= Hs <0.5', '0.5<= Hs <1.0',
                '1.0<= Hs <2.0', '2.0<= Hs <3.0',
                '3.0<= Hs <4.0', '4.0<= Hs <9.0']
spring_table.index = ['0<= θ <15', '15<= θ <30', '30<= θ <45', '45<= θ <60',
               '60<= θ <75', '75<= θ <90', '90<= θ <105', '105<= θ <120',
              '120<= θ <135', '135<= θ <150', '150<= θ <165', '165<= θ <180',
              '180<= θ <195', '195<= θ <210', '210<= θ <225', '225<= θ <240',
              '240<= θ <255', '255<= θ <270', '270<= θ <285', '285<= θ <300',
              '300<= θ <315', '315<= θ <330', '330<= θ <345', '345<= θ <360']
```

```python
# This line creates a new dataframe by selecting rows where the ANNO column is 1979, the MESE column is 1, the GIORNO column is 1, and the ORA column is 1
date_1 = data[data['ANNO']==1979][data['MESE']==1][data['GIORNO']==1][data['ORA']==1]

# This line creates a new dataframe by selecting rows where the ANNO column is 2020, the MESE column is 12, the GIORNO column is 31, and the ORA column is 23
date_2 = data[data['ANNO']==2020][data['MESE']==12][data['GIORNO']==31][data['ORA']==23]

# This line retrieves the integer index value of the first row of the date_1 and date_2 dataframes
ind_date_1 = int(date_1.index.values)
ind_date_2 = int(date_2.index.values)

# This line creates a new figure with a specified size
plt.figure(figsize=(20,5))

# This line creates a range of x values from the index of the first row of date_1 to the index of the last row of date_2
x = range(ind_date_1,ind_date_2)

# This line creates a y array from the Hs array that contains data for the time range of date_1 and date_2
y = Hs[ind_date_1:ind_date_2]

# This line sets a threshold value of 2.8
threshold = 2.8

# This line creates an array that contains the indices of the y values that are greater than or equal to the threshold
above_threshold = np.where(y >= threshold)

# This line creates an array that contains the minimum value between each y value and the threshold
below_threshold = np.minimum(y, threshold)

# This line plots the y values against the x values
plt.plot(x,y)

# This line creates a scatter plot of the indices of the y values that are above the threshold against their corresponding y values
plt.scatter(above_threshold, y[above_threshold], color= 'r')

plt.show()
```

```python
#This line returns the values of the array y that are above the threshold.
y[above_threshold]
```

```python
#These lines calculate the mean and standard deviation of the y values that are above the threshold, and print both of them to the console.
H_medio = np.mean(y[above_threshold])
H_std = np.std(y[above_threshold])
```

```python
H_medio, H_std
```

```python
#These lines calculate the values of theta and eps based on the mean and standard deviation calculated earlier, and print both of them to the console.
theta = H_std/1.2825
eps = H_medio - 0.45006*H_std
```

```python
theta, eps
```

```python
#These lines calculate the values of N and M based on the length of y and the difference between the maximum and minimum values in the yyyy array.
N = len(y[above_threshold]);
M = np.max(yyyy) - np.min(yyyy)

#These lines define an array Tr and calculate the values of Tr_star and P.
Tr = np.array([1, 5, 10, 25, 50, 100, 500])
Tr_star = Tr*(N/M)
P = 1 - (1/Tr_star)

#This line calculates the values of H_Tr based on eps, theta, and P.
H_Tr = eps + theta*(-np.log(-np.log(P)))
```

```python
#These lines create a plot of H_Tr vs. Tr, with a logarithmic x-axis, and add labels to the x-axis and y-axis. Finally, the plot is shown.

plt.figure(figsize=(7,5))
plt.semilogx(Tr,H_Tr)
plt.xlabel('T_r (year)', size=10)
plt.ylabel('H_s (m)', size=10)
plt.grid()
plt.show()
```

```python
# sort the significant wave heights in descending order
H_ord = np.sort(y[above_threshold])

# reverse the order of the array
H_ord = H_ord[::-1]

T_o = []
for i in range(1,len(H_ord)):
    F_i = 1-(i/(N+1))           # calculate non-exceedance probability
    Tr_ord = (1/(1-F_i))*M/N    # calculate the return period
    T_o.append(Tr_ord)
```

```python
plt.figure(figsize=(7,5))
plt.semilogx(Tr,H_Tr)                # plot the H_Tr curve
plt.semilogx(T_o,H_ord[1:])          # plot the return period vs H_s curve
plt.xlabel('T_r (year)', size=10)
plt.ylabel('H_s (m)', size=10)
plt.grid()
plt.show()
```

```python
P_H = np.exp(-(np.exp(-(y[above_threshold]-eps)/theta)))    # calculate the non-exceedance frequency
P_H_O = np.exp(-(np.exp(-(H_ord-eps)/theta)))               # calculate the non-exceedance frequency for H_ord
```
